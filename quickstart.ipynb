# FFT Quickstart

Code adapted from https://pythonnumericalmethods.studentorg.berkeley.edu/notebooks/chapter24.04-FFT-in-Python.html

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
SAMPLING_RATE = 1000
TIMESTEP = 1/SAMPLING_RATE
DURATION = 5

x = np.linspace(0,DURATION,DURATION*SAMPLING_RATE)
y = 10*np.sin((2*np.pi)*10*x) + 10*np.sin((2*np.pi)*50*x) + 10*np.sin((2*np.pi)*100*x) + 30*np.random.random(DURATION*SAMPLING_RATE)
```

```python
plt.plot(x,y)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("Signal in the Time-Domain")
```

```python
y_fft = np.fft.fft(y)
x_fft = np.fft.fftfreq(y_fft.shape[0], d=TIMESTEP)

plt.stem(x_fft, abs(y_fft))#, markerfmt=' ')
plt.xlim(0,)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.title("Frequency Domain")
# plt.ylim(0)
```
